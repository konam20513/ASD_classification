```python
from scipy.io import arff
import pandas as pd

# Replace 'your_file.arff' with the actual path to your ARFF file
file_path = 'data.arff'

# Load ARFF file
data, meta = arff.loadarff(file_path)

# Convert to pandas DataFrame and decode bytes to strings
df = pd.DataFrame(data)
df = df.applymap(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
```

```python
df
```

```python
# Rename the "Class" column to "label"
df.rename(columns={'Class/ASD': 'label'}, inplace=True)

# Replace "NO" with 0 and "YES" with 1 in the "label" column
df['label'] = df['label'].replace({'NO': 0, 'YES': 1})
```

```python
df
```

***Data Pre-processing***

```python
# Check for missing values or '?' symbols in all columns
for column in df.columns:
    missing_values = df[column].isin(['?', None]).sum()
    if missing_values > 0:
        print(f"Column '{column}' has {missing_values} missing values or '?' symbols.")
```

```python
# Check for missing values or '?' symbols in all columns
for column in df.columns:
    missing_values = df[column].isin(['?', None]).sum()
    nan_values = df[column].isnull().sum()
    
    if missing_values > 0 or nan_values > 0:
        print(f"Column '{column}' has {missing_values} missing values or '?' symbols and {nan_values} NaN values.")
```

```python
# Removed rows with ? and nan values
df_cleaned = df[~df.apply(lambda row: row.isin(['?', None]).any(), axis=1)].dropna()
```
