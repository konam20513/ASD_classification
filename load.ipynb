***Data loading***

```python
import pandas as pd
```

```python
df1=pd.read_csv('Toddler Autism dataset July 2018.csv')
df2=pd.read_csv('autism_screening.csv')
df3=pd.read_csv('data_csv.csv')
```

```python
dataframes = [df1, df2, df3]

# Lowercase column names for each DataFrame
for df in dataframes:
    df.columns = df.columns.str.lower()

# Print lists of columns for each DataFrame
for i, df in enumerate(dataframes, start=1):
    print(f"Columns for df{i}:")
    print(list(df.columns))
    print()
```

***Removing case_no from df1 and df3 as they are not present in df2***

```python
# Assuming df1 and df3 are your DataFrames
df1 = df1.drop(df1.columns[0], axis=1)  # Drop the first column in df1
df3 = df3.drop(df3.columns[0], axis=1)  # Drop the first column in df3
```

```python
dataframes = [df1, df2, df3]# Print lists of columns for each DataFrame
for i, df in enumerate(dataframes, start=1):
    print(f"Columns for df{i}:")
    print(list(df.columns))
    print()
```

***renaming first 10 columns in df2***

```python
new_df1 = df2.copy()

# Create a dictionary to map old column names to new column names for the first 10 columns
new_columns = {f'a{i}_score': f'a{i}' for i in range(1, 11)}

# Rename columns in the new DataFrame
new_df1.rename(columns=new_columns, inplace=True)

df2=new_df1
```

```python
new_df3 = df3.copy()

new_column_name = {'a10_autism_spectrum_quotient': 'a10'}

# Rename columns in the new DataFrame
new_df3.rename(columns=new_column_name, inplace=True)
df3=new_df3
```

```python


df3.rename(columns=new_column_name, inplace=True)
```

```python
dataframes = [df1, df2, df3]# Print lists of columns for each DataFrame
for i, df in enumerate(dataframes, start=1):
    print(f"Columns for df{i}:")
    print(list(df.columns))
    print()
```

***Creating single df***

```python
# Select the subset of columns from each dataframe
df1_subset = df1[['a1', 'a2', 'a3', 'a4', 'a5', 'a6', 'a7', 'a8', 'a9', 'a10', 'age_mons', 'sex', 'ethnicity', 'jaundice', 'family_mem_with_asd', 'who completed the test', 'class/asd traits ']]
df2_subset = df2[['a1', 'a2', 'a3', 'a4', 'a5', 'a6', 'a7', 'a8', 'a9', 'a10', 'age', 'gender', 'ethnicity', 'jundice', 'austim', 'relation', 'class/asd']]
df3_subset = df3[['a1', 'a2', 'a3', 'a4', 'a5', 'a6', 'a7', 'a8', 'a9', 'a10', 'age_years', 'sex', 'ethnicity', 'jaundice', 'family_mem_with_asd', 'who_completed_the_test', 'asd_traits']]
```

```python
# Rename columns in df1_subset and df3_subset to match df2_subset
rename_map_df1 = {
    'age_mons': 'age',
    'sex': 'gender',
    'jaundice': 'jundice',
    'family_mem_with_asd': 'austim',
    'who completed the test': 'relation',
    'class/asd traits ': 'class/asd'
}

rename_map_df3 = {
    'age_years': 'age',
    'sex': 'gender',
    'jaundice': 'jundice',
    'family_mem_with_asd': 'austim',
    'who_completed_the_test': 'relation',
    'asd_traits': 'class/asd'
}
```

```python
df1_subset = df1_subset.rename(columns=rename_map_df1)
df3_subset = df3_subset.rename(columns=rename_map_df3)

# Concatenate the dataframes
mega_df = pd.concat([df1_subset, df2_subset, df3_subset], ignore_index=True)
```

```python
mega_df
```

***Removing NaN vlaues***

```python
df=mega_df
```

```python
# Check for missing values or '?' symbols in all columns
for column in df.columns:
    missing_values = df[column].isin(['?', None]).sum()
    nan_values = df[column].isnull().sum()
    
    if missing_values > 0 or nan_values > 0:
        print(f"Column '{column}' has {missing_values} missing values or '?' symbols and {nan_values} NaN values.")
```

```python
# Removed rows with ? and nan values
df_cleaned = df[~df.apply(lambda row: row.isin(['?', None]).any(), axis=1)].dropna()
```

```python

# Rename the "Class" column to "label"
df_cleaned.rename(columns={'class/asd': 'label'}, inplace=True)
```

```python
df_cleand_lower = df_cleaned.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

***Replacing bi category columns with 0/1***

```python
# Replace "NO" with 0 and "YES" with 1 in the "label" column
df_cleand_lower['label'] = df_cleand_lower['label'].replace({'no': 0, 'yes': 1})
# Replace "NO" with 0 and "YES" with 1 in the "label" column
df_cleand_lower['austim'] = df_cleand_lower['austim'].replace({'no': 0, 'yes': 1})
df_cleand_lower['jundice'] = df_cleand_lower['jundice'].replace({'no': 0, 'yes': 1})
df_cleand_lower['gender'] = df_cleand_lower['gender'].replace({'f': 0, 'm': 1})
```

```python
# Assuming df_cleand is your DataFrame
df_cleand_lower['age'] = df_cleand_lower['age'].astype(int)
```
